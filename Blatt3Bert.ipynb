```python
#Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#Definiere Funktionen
def cosinus(n,x,w,t):
    return (np.cos(w*t-(w/c*n)*x))
    #c is global variable
def gauss(w0,w,sigma):
    return np.exp(-(((w-w0)/sigma)**2))
def n(wi):
    wertn= np.sqrt(1+rho/(wc**2-wi**2+ 0.000001))
    #workarbound to avoid ploblems with wc=wi
    return wertn
def wellenpaket(w0,sigma,x,t): 
    maxw=40
    wert=0
    nsigma=3
    for i in range(1, maxw):
        wi=w0-nsigma/2*sigma + nsigma*sigma/maxw*i
        ni= n(wi)
        # print (ni)
        wert= wert + gauss(w0,wi,sigma)*cosinus(ni,x,wi,t)
    return wert
```

```python
#Definiere Parameter 
c=1     # Geschwindigkeit
w0=10   # center frequence wave packet 
s=2     # sigma of frequency spectrum
wc=13   # resonance frequence medium for dispersion correction
#rho=0.0 # dispersion correction zero
rho=0.1 # positive value for anomalous  dispersion

#t=0  #Startzeit
t = 500  #späterer Zeitpunkt 

dw = 0.1  # diskrete Frequenzänderung pro Welle für Paket
w = np.arange(w0-s,w0+s,dw)  # Frequenzen im Paket
narray = n(w)
plottitle = 'Dispersion', t
fig, axes = plt.subplots()
axes.set_title(plottitle)
axes.set_xlabel("w")
axes.set_ylabel("n")
axes.plot(w,narray,linewidth=1)



#Erstelle Wellenpaket
dx = 0.001
x = np.arange(c*t-10,c*t+10,dx)
wellenzug = wellenpaket(w0,s,x,t)

#Plot
plottitle = 'Wellenzug, t= ', t
fig, axes = plt.subplots()
axes.set_title(plottitle)
axes.set_xlabel("x")
axes.set_ylabel("Amplitude")
axes.plot(x,wellenzug,linewidth=1)
```
