```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import matplotlib
from matplotlib import pyplot as plt
```

```python
r = 0.1
n_1 = 1
n_2 = 1.5
def f_b(h):
    return h/(np.tan(np.arcsin(h/r)-np.arcsin((n_1*h)/(n_2*r))))+r*(1-np.cos(np.arcsin(h/r)))
```

```python
x = np.arange(0, 0.099, 0.001)
fig, ax = plt.subplots()
plt.ylim(0, 0.5)
ax.plot(x, f_b(x))
ax.hlines(y=0.3, xmin=0.0, xmax=0.1)
plt.xlabel(' Abstand h[m]')
plt.ylabel('bildseitige Brennweite f_b[m]')
plt.title('Aufgabe 36')
plt.show() 
```
